## Subset Variables (Columns)

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
columns = ['sepal_width','sepal_length','species']
df[columns].head()
```

```python
df['sepal_width'].head()
```

```python
df.sepal_width.head()
```

```python
df.filter(regex='_').head()  # _값이 들어있는 컬럼을 가져옵니다.
```

```python
df.filter(regex='length$').head()  # length로 끝나는 컬럼을 가져옵니다.
```

```python
df.filter(regex='^sepal').head()  # sepal 문자로 시작하는 컬럼을 가져옵니다.
```

```python
df.filter(regex='^s[1-5]$').head()  # s로 시작하고 1~5로 끝나는 컬럼을 가져옵니다. (현재 데이터에는 존재하지 않습니다)
```

```python
df.filter(regex='^(?!species$).*').head()  # species 문자가 들어있지 않은 컬럼을 가져옵니다.
```

```python
df.loc[:5,'sepal_width':'petal_width']  # 5값을 포함합니다.
```

```python
df.iloc[:5,[1,2,4]]  # 5값을 포함하지 않습니다.
```

```python
df.loc[df['sepal_length'] > 7, ['sepal_length','sepal_width']]
```
