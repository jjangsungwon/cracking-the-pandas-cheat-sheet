```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
s = pd.Series(np.random.randn(1000),
             index=pd.date_range('1/1/2015', periods=1000))
```

```python
s = s.cumsum()
s.plot()
```

```python
r = s.rolling(window=30)  # window : 연산 갯수를 의미합니다.
r.mean()
```

```python
s.plot(style='k--')
r.mean().plot(style='k')
```

```python
df = pd.DataFrame(np.random.randn(1000, 4),
                 index=pd.date_range('1/1/2000', periods=1000),
                 columns=['A', 'B', 'C', 'D'])
```

```python
df = df.cumsum()
df.plot()
df.rolling(window=60).sum().plot(subplots=True)
```

```python
df.rolling(window=len(df), min_periods=1).mean().plot()
```

```python
df.expanding(min_periods=1).mean().plot()  # expanding은 window를 누적하면서 수치화하는 방법입니다.
```
